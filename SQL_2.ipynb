3. Try to join the actor, film_actor, and film table together (3 joins!)
SELECT film.file_id, file.title, film.description, film.release_year, file.language_id, film.rental_duration, film.rental_rate, film.length, film.replacement_cost, film.rating, film.last_update, film.special_features, film.fulltext, film_actor.actor_id, actor.first_name, actor.last_name

FROM film AS film

FULL JOIN

film_actor AS film_actor

ON film.film_id = film_actor.film_id

FULL JOIN

actor AS actor

ON film_actor.last_update = actor.last_update;
